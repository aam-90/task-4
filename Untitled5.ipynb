```python
# Problem 1: Mobile Recharge Validation System

def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ‚Çπ"))
        if amount < 50:
            print("Recharge amount must be at least ‚Çπ50. Try again.\n")
            continue
        if amount in valid_plans:
            print("Recharge Successful ‚úÖ")
            break
        else:
            print("Invalid plan. Available plans are:", valid_plans)
            print("Please try again.\n")
validate_recharge()
```

```python
# Problem 2: Inventory Reorder Alert System

def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(product, "‚Üí Reorder Alert ‚ö†")
        else:
            print(product, "‚Üí Stock OK ‚úÖ")
inventory = {
    "Shampoo": 10,
    "Soap": 25,
    "Toothpaste": 8,
    "Oil": 20
}
check_inventory(inventory)
```

```python
# Problem 3: Student Result Processing System

def process_result(marks):
    total = 0
    for mark in marks:
        total += mark
    average = total / len(marks)
    if average >= 50:
        return f"Average: {average:.2f} ‚Üí Pass ‚úÖ"
    else:
        return f"Average: {average:.2f} ‚Üí Fail ‚ùå"
student_marks = [60, 45, 70, 55]
print(process_result(student_marks))
```

```python
# Problem 4: Cab Fare Estimator with Retry Option

def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (12 * distance)
    if peak == "yes":
        fare += fare * 0.25  # 25% extra
    return fare
while True:
    km = float(input("Enter distance in km: "))
    peak_hour = input("Is it peak hour? (yes/no): ").lower()
    total_fare = calculate_fare(km, peak_hour)
    print(f"Total Fare: ‚Çπ{total_fare:.2f}")
    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        print("Thank you for using Cab Fare Estimator üöñ")
        break
```

```python
# Problem 5: Employee Attendance Eligibility Checker

def check_attendance(attendance_list):
    present_days = 0

    # Count present days
    for day in attendance_list:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance_list)) * 100

    if percentage >= 75:
        return f"Attendance: {percentage:.2f}% ‚Üí Eligible ‚úÖ"
    else:
        return f"Attendance: {percentage:.2f}% ‚Üí Not Eligible ‚ùå"

# Example
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
print(check_attendance(attendance))
```
